```python
inp_sample = """???.### 1,1,3
.??..??...?##. 1,1,3
?#?#?#?#?#?#?#? 1,3,1,6
????.#...#... 4,1,1
????.######..#####. 1,6,5
?###???????? 3,2,1"""
```

```python
from functools import cache

def solve(patterns, groups_list):
    ans = 0
    for i in range(len(patterns)):
        pattern = patterns[i] + ["."]
        groups = groups_list[i]
        
        @cache
        def solve1_single(curgroup, i, gi):
            if i == len(pattern):
                if gi == len(groups):
                    return 1
                return 0

            dot_sol = 0
            hash_sol = 0
            if pattern[i] == "?" or pattern[i] == ".":
                if curgroup == 0:
                    dot_sol = solve1_single(0, i + 1, gi)
                elif gi < len(groups) and curgroup == groups[gi]:
                    dot_sol = solve1_single(0, i + 1, gi + 1)
            if pattern[i] == "?" or pattern[i] == "#":
                if gi < len(groups) and curgroup + 1 <= groups[gi]:
                    hash_sol = solve1_single(curgroup + 1, i + 1, gi)

            if pattern[i] == "?":
                return dot_sol + hash_sol
            if pattern[i] == "#":
                return hash_sol
            if pattern[i] == ".":
                return dot_sol
            
        ans += solve1_single(0, 0, 0)
        
    return ans
        
def solve1(inp):
    rows = [s.split(" ") for s in inp.splitlines()]
    patterns = [list(row[0]) for row in rows]
    groups_list = [[int(x) for x in row[1].split(",")] for row in rows]
    return solve(patterns, groups_list)
```

```python
def solve2(inp):
    rows = [s.split(" ") for s in inp.splitlines()]
    patterns = [list("?".join([row[0]]*5)) for row in rows]
    groups = [[int(x) for x in row[1].split(",")]*5 for row in rows]
    return solve(patterns, groups)
```
