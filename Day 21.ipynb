```python
inp_sample = """...........
.....###.#.
.###.##..#.
..#.#...#..
....#.#....
.##..S####.
.##..#...#.
.......##..
.##.#.####.
.##..##.##.
..........."""
```

```python
neighbors = [(-1, 0), (1, 0), (0, 1), (0, -1)]
```

```python
from collections import deque

def solve1(inp, steps=64):
    inp = [list(s) for s in inp.split('\n')]
    n, m = len(inp), len(inp[0])
    start = None
    for i in range(n):
        for j in range(m):
            if inp[i][j] == "S":
                start = (i, j)
                inp[i][j] = "."
    
    
    reachable_spots = set()
    
    q = deque()
    q.append(start)

    for step in range(steps):
        prev_frontier_len = len(q)
        this_time = set()
        while len(q) > 0:
            i, j = q.popleft()
            reachable_spots.add((i, j))
            for di, dj in neighbors:
                ni, nj = i + di, j + dj
                if 0 <= ni < n and 0 <= nj < m and inp[ni][nj] == "." and (ni, nj) not in this_time:
                    this_time.add((ni, nj))
        for x in this_time:
            q.append(x)
        if len(q) + prev_frontier_len == len(reachable_spots):
            if steps % 2 != step % 2:
                return len(q)
            else:
                return prev_frontier_len
    return len(q)
```

```python
from collections import deque

def solve2(inp, steps=26501365):
    inp = [list(s) for s in inp.split('\n')]
    n, m = len(inp), len(inp[0])
    start = None
    for i in range(n):
        for j in range(m):
            if inp[i][j] == "S":
                start = (i, j)
                inp[i][j] = "."

    q = deque()
    q.append(start)
    for _ in range(steps):
        this_time = set()
        while len(q) > 0:
            i, j = q.popleft()
            for di, dj in neighbors:
                ni, nj = i + di, j + dj
                if 0 <= ni < n and 0 <= nj < m and inp[ni][nj] == "." and (ni, nj) not in this_time:
                    this_time.add((ni, nj))
        for x in this_time:
            q.append(x)
    return len(q)
```
