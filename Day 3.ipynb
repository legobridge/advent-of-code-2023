```python
not_symbols = digits + '.'
print(not_symbols)
```

```python
neighbors_deltas = [(i, j) for i in range(-1, 2) for j in range(-1, 2)]
neighbors_deltas.remove((0, 0))
neighbors_deltas
```

```python
def solve(inp):
    ans = 0
    st = inp.split('\n')
    n, m = len(st), len(st[0])
    for i in range(n):
        for j in range(m):
            if st[i][j] not in digits:
                continue
            if j > 0 and st[i][j - 1] in digits:
                continue
            tj = j
            num = 0
            symbol_found = False
            while tj < m and st[i][tj] in digits:
                num = num * 10 + int(st[i][tj])
                for nd in neighbors_deltas:
                    ni, nj = i + nd[0], tj + nd[1]
                    if 0 <= ni < n and 0 <= nj < m and st[ni][nj] not in not_symbols:
                        symbol_found = True
                        break
                tj += 1
            if symbol_found:
                ans += num
    return ans
```

```python
def solve2(inp):
    ans = 0
    st = inp.split('\n')
    n, m = len(st), len(st[0])
    gears = [[[] for _ in range(m)] for _ in range(n)]
    for i in range(n):
        for j in range(m):
            if st[i][j] not in digits:
                continue
            if j > 0 and st[i][j - 1] in digits:
                continue
            tj = j
            num = 0
            gearset = set()
            while tj < m and st[i][tj] in digits:
                num = num * 10 + int(st[i][tj])
                for nd in neighbors_deltas:
                    ni, nj = i + nd[0], tj + nd[1]
                    if 0 <= ni < n and 0 <= nj < m and st[ni][nj] == "*":
                        gearset.add((ni, nj))
                tj += 1
            for ni, nj in gearset:
                gears[ni][nj].append(num)

    for gear_list in gears:
        for gear in gear_list:
            if len(gear) == 2:
                ans += gear[0] * gear[1]
    return ans
```
