```python
neighbors = [(-1, 0), (0, -1), (1, 0), (0, 1)]
```

```python
from collections import deque

def solve1(inp, expansion=2):
    rows = [list(s) for s in inp.splitlines()]
    n, m = len(rows), len(rows[0])
    
    row_coordinates = [0]
    for i in range(n):
        non_empty = False
        for j in range(m):
            if rows[i][j] != '.':
                non_empty = True
                break
        if non_empty:
            row_coordinates.append(row_coordinates[-1] + 1)
        else:
            row_coordinates.append(row_coordinates[-1] + expansion)
            
    col_coordinates = [0]
    for j in range(m):
        non_empty = False
        for i in range(n):
            if rows[i][j] != '.':
                non_empty = True
                break
        if non_empty:
            col_coordinates.append(col_coordinates[-1] + 1)
        else:
            col_coordinates.append(col_coordinates[-1] + expansion)
            
    
    galaxies = []
    for i in range(n):
        for j in range(m):
            if rows[i][j] == "#":
                galaxies.append((i, j))
    
    # print(galaxies)
    ans = 0
    for gi1 in range(len(galaxies)):
        for gi2 in range(gi1 + 1, len(galaxies)):
            xdist = row_coordinates[galaxies[gi1][0]] - row_coordinates[galaxies[gi2][0]]
            ydist = col_coordinates[galaxies[gi1][1]] - col_coordinates[galaxies[gi2][1]]
            ans += abs(xdist) + abs(ydist)
    return ans
```

```python
solve1(inp_sample)
```

```python
solve1(inp)
```

```python
def solve2(inp):
    return solve1(inp, 1000000)
```
