```python
inp_sample = r""".|...\....
|.-.\.....
.....|-...
........|.
..........
.........\
..../.\\..
.-.-/..|..
.|....-|.\
..//.|...."""
```

```python
from collections import deque

def solve1(inp, start=(0, 0, 0, 1)):
    rows = [list(s) for s in inp.splitlines()]
    n, m = len(rows), len(rows[0])
    squares = set()
    visited = set()
    q = deque()
    q.append(start)
    while len(q) > 0:
        i, j, di, dj = q.popleft()
        if not (0 <= i < n and 0 <= j < m):
            continue
        squares.add((i, j))
        visited.add((i, j, di, dj))
        neighbors = []
        if rows[i][j] == '.':
            neighbors.append((i + di, j + dj, di, dj))
        elif rows[i][j] == '/':
            neighbors.append((i - dj, j - di, -dj, -di))
        elif rows[i][j] == '\\':
            neighbors.append((i + dj, j + di, dj, di))
        elif rows[i][j] == '|':
            if (di, dj) == (0, 1) or (di, dj) == (0, -1):
                neighbors.append((i + 1, j, 1, 0))
                neighbors.append((i - 1, j, -1, 0))
            else:
                neighbors.append((i + di, j + dj, di, dj))
        elif rows[i][j] == '-':
            if (di, dj) == (1, 0) or (di, dj) == (-1, 0):
                neighbors.append((i, j + 1, 0, 1))
                neighbors.append((i, j - 1, 0, -1))
            else:
                neighbors.append((i + di, j + dj, di, dj))
        for neighbor in neighbors:
            if neighbor not in visited:
                q.append(neighbor)
    return len(squares)
```

```python
solve1(inp_sample)
```

```python
solve1(inp)
```

```python
def solve2(inp):
    rows = [list(s) for s in inp.splitlines()]
    n, m = len(rows), len(rows[0])
    ans = 0
    for j in range(m):
        ans = max(ans, solve1(inp, (0, j, 1, 0)))
        ans = max(ans, solve1(inp, (n - 1, j, -1, 0)))
    for i in range(n):
        ans = max(ans, solve1(inp, (i, 0, 0, 1)))
        ans = max(ans, solve1(inp, (i, m -1, 0, -1)))
    return ans
```

```python
solve2(inp)
```
