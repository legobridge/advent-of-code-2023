```python
def get_brick_spots(start, end):
    spots = set()
    for i in range(start[0], end[0] + 1):
        for j in range(start[1], end[1] + 1):
            for k in range(start[2], end[2] + 1):
                spots.add((i, j, k))
    return spots

def solve1(inp):
    bricks = [s.split("~") for s in inp.split('\n')]
    n = len(bricks)
    for i in range(n):
        bricks[i] = [[int(x) for x in s.split(",")] for s in bricks[i]]
    
    # print(bricks)
    bricks.sort(key=lambda x: x[0][2])
    filled = set()
    for start, end in bricks:
        for spot in get_brick_spots(start, end):
            filled.add(spot)
    
    # print(bricks)
    # print(sorted(filled))
    no_change = False
    while not no_change:
        no_change = True
        for start, end in bricks:
            if start[2] == 1:
                continue
            new_spots_start = (start[0], start[1], start[2] - 1)
            new_spots_end = (end[0], end[1], start[2] - 1)
            new_spots = get_brick_spots(new_spots_start, new_spots_end)
            if len(new_spots.intersection(filled)) == 0:
                filled = filled | new_spots
                old_spots_start = (start[0], start[1], end[2])
                old_spots_end = (end[0], end[1], end[2])
                old_spots = get_brick_spots(old_spots_start, old_spots_end)
                filled = filled - old_spots
                start[2] -= 1
                end[2] -= 1
                no_change = False
    
    # print(bricks)
    # print(sorted(filled))
    
    filled = {}
    for i in range(n):
        start, end = bricks[i]
        for spot in get_brick_spots(start, end):
            filled[spot] = i
    
    necessary = set()
    for i in range(n):
        start, end = bricks[i]
        if start[2] == 1:
            continue
        bottom_spots_start = (start[0], start[1], start[2] - 1)
        bottom_spots_end = (end[0], end[1], start[2] - 1)
        bottom_spots = get_brick_spots(bottom_spots_start, bottom_spots_end)
        my_supps = set()
        for bs in bottom_spots:
            if bs in filled:
                my_supps.add(filled[bs])
        if len(my_supps) == 1:
            necessary |= my_supps
    
    return n - len(necessary)
```

```python
from collections import defaultdict

def get_brick_spots(start, end):
    spots = set()
    for i in range(start[0], end[0] + 1):
        for j in range(start[1], end[1] + 1):
            for k in range(start[2], end[2] + 1):
                spots.add((i, j, k))
    return spots

def disintegrate(brick, i_supp, in_degree):
    ans = 0
    for v in i_supp[brick]:
        in_degree[v] -= 1
        if in_degree[v] == 0:
            ans += 1 + disintegrate(v, i_supp, in_degree)
    return ans

def solve2(inp):
    bricks = [s.split("~") for s in inp.split('\n')]
    n = len(bricks)
    for i in range(n):
        bricks[i] = [[int(x) for x in s.split(",")] for s in bricks[i]]
    
    # print(bricks)
    bricks.sort(key=lambda x: x[0][2])
    filled = set()
    for start, end in bricks:
        for spot in get_brick_spots(start, end):
            filled.add(spot)
    
    # print(bricks)
    # print(sorted(filled))
    no_change = False
    while not no_change:
        no_change = True
        for start, end in bricks:
            if start[2] == 1:
                continue
            new_spots_start = (start[0], start[1], start[2] - 1)
            new_spots_end = (end[0], end[1], start[2] - 1)
            new_spots = get_brick_spots(new_spots_start, new_spots_end)
            if len(new_spots.intersection(filled)) == 0:
                filled = filled | new_spots
                old_spots_start = (start[0], start[1], end[2])
                old_spots_end = (end[0], end[1], end[2])
                old_spots = get_brick_spots(old_spots_start, old_spots_end)
                filled = filled - old_spots
                start[2] -= 1
                end[2] -= 1
                no_change = False
    
    # print(bricks)
    # print(sorted(filled))
    
    filled = {}
    for i in range(n):
        start, end = bricks[i]
        for spot in get_brick_spots(start, end):
            filled[spot] = i
    
    necessary = set()
    i_supp = defaultdict(list)
    in_degree = [0 for _ in range(n)]
    for i in range(n):
        start, end = bricks[i]
        if start[2] == 1:
            continue
        bottom_spots_start = (start[0], start[1], start[2] - 1)
        bottom_spots_end = (end[0], end[1], start[2] - 1)
        bottom_spots = get_brick_spots(bottom_spots_start, bottom_spots_end)
        my_supps = set()
        for bs in bottom_spots:
            if bs in filled:
                my_supps.add(filled[bs])
        for my_supp in my_supps:
            i_supp[my_supp].append(i)
            in_degree[i] += 1
        if len(my_supps) == 1:
            necessary.add(my_supps.pop())
    
    ans = 0
    for brick in necessary:
        ans += disintegrate(brick, i_supp, in_degree.copy())
    return ans
```
