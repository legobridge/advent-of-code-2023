```python
def get_area_from_vertices(vertices):
    interior_area = 0
    boundary_points = 4
    for i in range(len(vertices) - 1):
        interior_area += (vertices[i][0] * vertices[i + 1][1] - vertices[i][1] * vertices[i + 1][0])
        boundary_points += abs(vertices[i][0] - vertices[i + 1][0]) + abs(vertices[i][1] - vertices[i + 1][1])
    interior_area = abs(interior_area // 2)
    return interior_area + (boundary_points // 2) - 1

def solve1(inp):
    rows = [s.split(" ") for s in inp.splitlines()]
    for row in rows:
        row[1] = int(row[1])
    vertices = [(0, 0)]
    for row in rows:
        new_vertex = list(vertices[-1])
        if row[0] == "R":
            new_vertex[1] += row[1]
        if row[0] == "L":
            new_vertex[1] -= row[1]
        if row[0] == "D":
            new_vertex[0] += row[1]
        if row[0] == "U":
            new_vertex[0] -= row[1]
        vertices.append(tuple(new_vertex))
        
    return get_area_from_vertices(vertices)
    
```

```python
def solve2(inp):
    rows = [s.split(" ") for s in inp.splitlines()]
    for row in rows:
        row[0] = row[2][-2]
        row[1] = int(row[2][2:-2], 16)
    vertices = [(0, 0)]
    for row in rows:
        new_vertex = list(vertices[-1])
        if row[0] == "0":
            new_vertex[1] += row[1]
        if row[0] == "2":
            new_vertex[1] -= row[1]
        if row[0] == "1":
            new_vertex[0] += row[1]
        if row[0] == "3":
            new_vertex[0] -= row[1]
        vertices.append(tuple(new_vertex))
        
    return get_area_from_vertices(vertices)
```
